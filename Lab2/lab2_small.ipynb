```
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import matplotlib.pyplot as plt
import numpy as np


# ======================
# LOAD SHAPE
# ======================
shp_path = "/content/uttaradit.shp"
gdf = gpd.read_file(shp_path)


# ======================
# LOAD RASTER
# ======================
raster_path = "/content/drive/MyDrive/data/2026-02-16-00_00_2026-02-16-23_59_Sentinel-2_L2A_True_color.tiff"

with rasterio.open(raster_path) as src:

    print("Raster CRS:", src.crs)
    print("Raster bounds:", src.bounds)

    # แปลง CRS ให้ตรง
    gdf = gdf.to_crs(src.crs)

    print("Shape bounds:", gdf.total_bounds)

    # Clip (สำคัญ)
    out_image, out_transform = mask(
        src,
        gdf.geometry,
        crop=True,
        all_touched=True   # ⭐ ตัวช่วยสำคัญ
    )

    print("Min Max:", out_image.min(), out_image.max())


# ======================
# RGB
# ======================
rgb = np.dstack((out_image[0], out_image[1], out_image[2])).astype(float)

# normalize
rgb_min = np.nanmin(rgb)
rgb_max = np.nanmax(rgb)

if rgb_max != rgb_min:
    rgb = (rgb - rgb_min) / (rgb_max - rgb_min)


# ======================
# Plot
# ======================
fig, ax = plt.subplots(figsize=(8,8))

ax.imshow(rgb)

gdf.boundary.plot(ax=ax, edgecolor="red", linewidth=2)

plt.title("Clipped Sentinel - Uttaradit")

plt.axis("off")

plt.show()
```
